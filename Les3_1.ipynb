```python
#Exersice 3

#3_1 Circle
r = 6
x = np.linspace(-r,r,1000)
y = np.sqrt(-x**2+r**2)
plt.plot(x, y,'b')
plt.plot(x,-y,'b')
plt.gca().set_aspect('equal')
plt.show()

```

```python
#Exersice 3

#3_2 Ellipse
a = 4
b = 2
x = np.linspace(-4,5,1000)
y = b * np.sqrt(1 - (x**2/a**2))
plt.plot(x, y,'b')
plt.plot(x,-y,'b')

plt.show()
```

```python
#Exersice 3

#3_3 Hyperbola
a = 1
b = 1
x = np.linspace(-10,10,1000)
y = b * np.sqrt( (x**2/a**2) - 1)
plt.plot(x, y,'b')
plt.plot(x,-y,'b')

plt.show()
```

```python
#Exersice 5

#5_1
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y + 100
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```

```python
#Exersice 5

#5_2
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
a = 5
b = 3
X, Y = np.meshgrid(X, Y)
Z = ((X**2)/a**2) + ((Y**2)/b**2)
Z2 = ((X**2)/a**2) + ((Y**2)/b**2) + 100
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```
