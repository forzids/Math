```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#y(x) = k∙cos(x – a) + b
def f(x):
    k = 2
    a = 3
    b = 4
    k1 = 3
    a1 = 5
    b1 = 0
    return k*np.cos(x-a)+b;k1*np.cos(x-a1)+b1
x = np.linspace(0, 5, 50)  
x1 = np.linspace(2, 10, 50)
y = f(x)
y1 = f(x1)
plt.plot(x, y)
plt.plot(x1, y1)
plt.show()
```

```python
#polar to cartesian
def polar2cart(r,phi):
    x = r*math.cos(phi)
    y = r*math.sin(phi)
    return x,y

polar2cart(1,10)
```

```python
#circle in polar
ang = np.arange(0., 2., 1./180.)*np.pi 

plt.polar(ang, [10]*len(ang)) 

plt.show()


```
